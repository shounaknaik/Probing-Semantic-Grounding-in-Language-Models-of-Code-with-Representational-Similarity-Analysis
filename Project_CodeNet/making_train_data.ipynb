```python
import glob
import pandas as pd
import random
import pickle

from tqdm import tqdm
```

```python
file_list=glob.glob('./metadata/p0*.csv')
```

```python
with open('test_problems.pkl','rb') as f:
    ok_problems=pickle.load(f)
```

```python
file_list = [x for x in file_list if x not in ok_problems]
```

```python
len(file_list)
```

```python
codebert_languages=['Python', 'Java', 'JavaScript', 'PHP', 'Ruby', 'Go']
```

```python
check_li=[(x, 'Accepted') for x in codebert_languages]
```

```python
check_li
```

```python
def create_dict():
    my_dict={}
    for tup in check_li:
        my_dict[tup]=0
    return my_dict
```

```python
problem_test_id=[]
```

```python
for file in tqdm(file_list,total=len(file_list)):
    
    df=pd.read_csv(file)
    df=df.loc[df['language'].isin(codebert_languages)]
    df=df.loc[df['status']== 'Accepted']
    
    my_dict=create_dict()
    
    for _ , row in df.iterrows():
        my_dict[(row['language'],row['status'])]+=1
    
    flag=1
    for value in my_dict.values():
        if value ==0:
            flag=0
            
    if flag:
        problem_test_id.append(file)
```

```python
len(problem_test_id)
```

```python
problem_test_id
```

```python
with open('./train_problems.pkl','wb') as f:
    pickle.dump(problem_test_id,f)
```

```python
final_df=pd.DataFrame()
```

```python
for file in tqdm(problem_test_id,total=len(problem_test_id)):
    df=pd.read_csv(file)
    final_df=pd.concat([final_df,df])
```

```python
final_df=final_df.loc[final_df['language'].isin(codebert_languages)]
```

```python
final_df=final_df.loc[final_df['status']=='Accepted']
```
