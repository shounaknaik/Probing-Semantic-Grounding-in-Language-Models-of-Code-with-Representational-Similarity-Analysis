```python
data_df = pd.read_json('/content/drive/MyDrive/Probing CodeBERT /test.jsonl', lines=True)
```

```python
def make_glove_dict():
  embeddings_dict = {}
  with open("/content/drive/MyDrive/Probing CodeBERT /glove.6B.300d.txt", 'r') as f:
    for line in tqdm(f,total=400001):
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], "float32")
        embeddings_dict[word] = vector
    pickle.dump(embeddings_dict,open('/content/drive/MyDrive/Probing CodeBERT /glove_embeddings_dict.pkl','wb'))
    return embeddings_dict
```

```python
req_dict=make_glove_dict()
```

```python
def get_glove_mean_embedding(sentence):
  unseen_words=[]
  li=[]
  for word in sentence.split():
    word=word.split('.')[0].lower()
    try:
      embed=req_dict[word]
      li.append(embed)
    except:
      unseen_words.append(word)
      continue
  mean_embeddings=np.mean(li,axis=0)
  return mean_embeddings,unseen_words
```

```python
def get_all_glove_embeddings():
  unseen_words_li=[]
  ser=data_df["docstring_summary"]
  for index, sentence in tqdm(enumerate(ser),total=len(ser)):
    embed,unseen_words=get_glove_mean_embedding(sentence)
    unseen_words_li.extend(unseen_words)
    # print(embed)
    with open(f'/content/drive/MyDrive/Probing CodeBERT /Glove_embeddings/line_{index}.npy', 'wb') as f:
      np.save(f, embed)
  
  return unseen_words_li

    

```
